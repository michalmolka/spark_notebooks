```
dbutils.widgets.text('iowa_id', '')
dbutils.widgets.text('iowa_name', '')
dbutils.widgets.text('iowa_value', '')
```

```
iowa_id = dbutils.widgets.get('iowa_id')
iowa_name = dbutils.widgets.get('iowa_name')
iowa_value = dbutils.widgets.get('iowa_value')

print(iowa_id)
print(iowa_name)
print(iowa_value)
```

```
spark.sql(f"""
  INSERT INTO iowa.adf_parameter 
  (id, name, value) 
  VALUES ({iowa_id}, '{iowa_name}',{iowa_value})
""")
```
